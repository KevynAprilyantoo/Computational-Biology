```python
import pandas as pd
from Bio import SeqIO
from Bio import Seq
```

```python
parseReadFile = SeqIO.parse("sequence.fasta","fasta")
print(parseReadFile)
seq=""
for n in parseReadFile:
    seq=n.seq
print(seq)
```

```python
readFile = SeqIO.read("sequence.fasta", "fasta")
print(readFile)
```

```python
seqA = readFile.seq
print(seqA)
seqA_rna = seqA.transcribe()
print(seqA_rna)
```
