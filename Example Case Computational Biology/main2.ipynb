```python
from Bio.Seq import Seq
from Bio import SeqIO
```

```python
readFileX = SeqIO.parse('x.fasta','fasta')
SeqX = ""
for n in readFileX:
    SeqX = n.seq
print(SeqX)
```

```python
SeqX = str(SeqX)
SeqX = SeqX.replace("<letter>","")
SeqX = Seq(SeqX)
```

```python
readFileY = SeqIO.parse('y.fasta','fasta')
SeqY = ""
for n in readFileY:
    SeqY = n.seq
print(SeqY)
```

```python
SeqX_length = len(SeqX)
SeqY_length = len(SeqY)

print(f"Total length of Sequence X : {SeqX_length}")
print(f"Total length of Sequence Y : {SeqY_length}")
```

```python
print(f"The fiest time the codon (CGC) appears in Sequence X : {SeqX.find('CGC')}")
print(f"The fiest time the codon (CGC) appears in Sequence Y : {SeqY.find('CGC')}")
```

```python
Seq1 = SeqX[:10] + SeqX[-5:]
print(Seq1)
Seq2 = SeqY[:20] + SeqX[-3:]
print(Seq2)
SeqZ = Seq1 + Seq2
SeqZ = SeqZ[::-1]
print(f"Sequence Z : {SeqZ}")
```

```python
import matplotlib.pyplot as plt
SeqZ_baseA = SeqZ.count('A')
SeqZ_baseC = SeqZ.count('C')

plt.bar("A", SeqZ_baseA)
plt.bar("C", SeqZ_baseC)
plt.show()

```

```python
from Bio.SeqUtils import gc_fraction

SeqZ_GC = gc_fraction(SeqZ)
SeqZ_AT = 1-SeqZ_GC

print(f"Sequence Z GC Percentage : {SeqZ_GC}")
print(f"Sequence Z AC Percentage : {SeqZ_AT}")

plt.title("Nucleotides Base A And C in Sequence Z")
plt.bar("GC", SeqZ_GC)
plt.bar("AT", SeqZ_AT)
plt.show
```

```python
from Bio.SeqUtils import MeltingTemp

SeqZ_MeltingTemp = MeltingTemp.Tm_Wallace(SeqZ)
print(f"Sequence Z Melting Temperature : {SeqZ_MeltingTemp}")
```

```python
from Bio.SeqUtils import molecular_weight

SeqZ_Molecular = molecular_weight(SeqZ)
print(f"Sequence Z Molecular Weight : {SeqZ_Molecular}")
```

```python
mRNA_SeqZ = SeqZ.transcribe()
print(f"Sequence Z mRNA : {mRNA_SeqZ}")
```

```python
from Bio.SeqUtils import seq3
SeqZ_protein = mRNA_SeqZ.translate()
print(SeqZ_protein)

SeqZ_proteinSeq = seq3(SeqZ_protein)
print("Sequence Z Asam Amino : " + SeqZ_proteinSeq)
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

global_alignment = pairwise2.align.globalxx(SeqX, SeqY)

for i in global_alignment:
    print(format_alignment(*i))

print("--------------------------------------------------------------------------------------")
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

local_alignment = pairwise2.align.localxx(SeqX, SeqY)

for i in local_alignment:
    print(format_alignment(*i))

print("--------------------------------------------------------------------------------------")
```

```python
def hamming_distance(seqA, seqB):
    count = 0
    for i, j in zip(seqA,seqB):
        if i!= j:
            count+=1
    return count

hamming_distance(SeqX, SeqY)
```

```python
import Levenshtein

SeqZ_Levenshtein = Levenshtein.distance(SeqX, SeqY)
print(SeqZ_Levenshtein)
```
