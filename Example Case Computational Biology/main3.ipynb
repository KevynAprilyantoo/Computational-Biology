```python
from Bio.Seq import Seq
from Bio import SeqIO
```

```python
readFileX = SeqIO.parse('x.fasta','fasta')
SeqX = ''
for n in readFileX:
    SeqX = n.seq
print(f'Sequence X : {SeqX}')
```

```python
readFileY = SeqIO.parse('y.fasta','fasta')
SeqY = ''
for n in readFileY:
    SeqY = n.seq
print(f'Sequence Y : {SeqY}')
```

```python
SeqX = str(SeqX)
SeqX = SeqX.replace("<letter>","Z")
SeqX = Seq(SeqX)
print(SeqX)
```

```python
print(f"The total length of Sequence X : {len(SeqX)}")
print(f"The total length of Sequence Y : {len(SeqY)}")
```

```python
print(f'Index the codon (CGC) appears in Sequence X : {SeqX.find("CGC")}')
print(f'Index the codon (CGC) appears in Sequence Y : {SeqY.find("CGC")}')
```

```python
Seq1 = SeqX[:10] + SeqX[-5:]
Seq2 = SeqY[:20] + SeqY[-3:]
SeqZ = Seq1 + Seq2
SeqZ = SeqZ[::-1]
print(f'Sequence Z : {SeqZ}')
```

```python
import matplotlib.pyplot as plt

SeqZ_A = SeqZ.count('A')
SeqZ_C = SeqZ.count('C')

plt.title("Nucleotides Base A and C Frequency")
plt.bar("A",SeqZ_A)
plt.bar("C", SeqZ_C)
plt.show()
```

```python
from Bio.SeqUtils import gc_fraction

SeqZ_GC = gc_fraction(SeqZ)
SeqZ_AT = 1-SeqZ_GC

print(f"GC percentage for Sequence Z : {SeqZ_GC}")
print(f"AT percentage for Sequence Z : {SeqZ_AT}")
```

```python
from Bio.SeqUtils import MeltingTemp
from Bio.SeqUtils import molecular_weight

SeqZ_melting = MeltingTemp.Tm_Wallace(SeqZ)
print(f"Sequence Z Melting Time : {SeqZ_melting}")

SeqZ_molecular = molecular_weight(SeqZ)
print(f"Sequence Z Molecular Weight : {SeqZ_molecular}")

```
